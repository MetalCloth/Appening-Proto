```python
table_queries = [
    "Compare LLMs and agents in a table",
    "What are the types of AI with their capabilities?",
    "Industry readiness parameters for manufacturing vs healthcare",
    "LLM vs agent interactivity and decision making"
]

for q in table_queries:
    results = vector_store.similarity_search_with_score(q, k=3)
    
    print(f"\n{'='*60}")
    print(f"Query: {q}")
    print('='*60)
    
    for doc, score in results:
        print(f"\nScore: {score:.4f}")
        print(f"Page: {doc.metadata.get('page', 'N/A')}")
        print(f"Text preview:\n{doc.page_content[:300]}...")
        
        # Check if looks like table content
        has_pipes = '|' in doc.page_content
        has_columns = any(x in doc.page_content.lower() for x in ['type', 'aspect', 'parameter', 'capability'])
        
        if has_pipes or has_columns:
            print("📊 TABLE DETECTED")
```
